```python
import numpy as np
import pandas as pd
```

```python
dataset_train = pd.read_csv('E:\\MASABAH\\MS DS 2k21\\Machine Learning\\Assignment1\\TrainingSet.csv')
dataset_test = pd.read_csv('E:\\MASABAH\\MS DS 2k21\\Machine Learning\\Assignment1\\TestSet1.csv')
```

```python
dataset_train.describe()
```

```python
feature_columns = ['leaf.length','leaf.width','flower.length','flower.width']

X_train = dataset_train[feature_columns].values

y_train = dataset_train['plant'].values

X_test = dataset_test[feature_columns].values

y_test = dataset_test['plant'].values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

plt.figure(figsize=(12, 6))
plt.title('Leaf Dimensions')

sns.scatterplot(x=dataset_train["leaf.length"], 
                y=dataset_train["leaf.width"], 
                hue=dataset_train["plant"],palette = ["green","orange","blue"],
                s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Flower Dimensions')

sns.scatterplot(x=dataset_train["flower.length"], 
                y=dataset_train["flower.width"], 
                hue=dataset_train["plant"],palette = ["green","orange","blue"],
                s=100);
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

#plot for leaf length
axes[0,0].set_title('Distribution of Leaf Length')
axes[0,0].hist(dataset_train["leaf.length"]);

#plot for leaf width
axes[0,1].set_title('Distribution of Leaf Width')
axes[0,1].hist(dataset_train["leaf.width"]);

#plot for flower length
axes[1,0].set_title('Distribution of Flower Length')
axes[1,0].hist(dataset_train["flower.length"]); 

#plot for flower width
axes[1,1].set_title('Distribution of Flower Width')
axes[1,1].hist(dataset_train["flower.width"]);
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(dataset_train.corr(), annot=True, cmap='Blues');
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

dataset_test['plant']=y_pred
dataset_test
```
